```python
data['RoP_cut']= pd.qcut(data['rate_of_penetration'], q=4, labels = ['0-25', '25-50', '50-75','75-100'])
data.drop(['segment_id', 'wellbore_chev_no_id'], axis='columns', inplace = True)
data.head()
```

```python
target = data.RoP_cut
inputs = data.drop(['RoP_cut'], axis = 'columns')
dummies = pd.get_dummies(inputs.formation_id)
inputs = pd.concat([inputs,dummies], axis = 'columns')
inputs.drop('formation_id', axis = 'columns', inplace = True)
```

```python
inputs.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size= 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
model.score(X_test, y_test)
```
